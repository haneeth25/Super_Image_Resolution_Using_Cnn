```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
DirPath_lr = '/content/drive/MyDrive/Data/Train/LR'
DirPath_hr = '/content/drive/MyDrive/Data/Train/HR'
img_names = os.listdir(DirPath_lr)
train_lr = []
train_hr = []

for each_img in img_names:
  img_path = os.path.join(DirPath_lr,each_img)
  img = cv2.imread(img_path)
  resized_img = cv2.resize(img,(224,224)) 
  train_lr.append(resized_img)

for each_img in img_names:
  img_path = os.path.join(DirPath_hr,each_img)
  img = cv2.imread(img_path)
  resized_img = cv2.resize(img,(224,224)) 
  train_hr.append(resized_img)
```

```python
DirPath_lr_val = '/content/drive/MyDrive/Data/Validation/LR'
DirPath_hr_val = '/content/drive/MyDrive/Data/Validation/HR'
img_names = os.listdir(DirPath_lr_val)
train_lr_val = []
train_hr_val = []

for each_img in img_names:
  img_path = os.path.join(DirPath_lr_val,each_img)
  img = cv2.imread(img_path)
  resized_img = cv2.resize(img,(224,224)) 
  train_lr_val.append(resized_img)

for each_img in img_names:
  img_path = os.path.join(DirPath_hr_val,each_img)
  img = cv2.imread(img_path)
  resized_img = cv2.resize(img,(224,224)) 
  train_hr_val.append(resized_img)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Conv2D,MaxPooling2D,Dense,Flatten,Dropout,UpSampling2D,Add
from keras.optimizers import Adam 
from keras.models import Model
```

```python
def Super_Resolution():
    input_img = Input(shape=train_lr.shape[1:])
    
    l1 = Conv2D(64,kernel_size=3,activation="relu",padding="same")(input_img)
    m1 = MaxPooling2D(pool_size=2)(l1)

    l2 = Conv2D(128,kernel_size=3,activation="relu",padding="same")(m1)
    m2 = MaxPooling2D(pool_size=2)(l2)
    
    l3 = Conv2D(256,kernel_size=3,activation="relu",padding="same")(m2)
    
    u1 = UpSampling2D(size=2)(l3)
    l4 = Conv2D(128,kernel_size=3,activation="relu",padding="same")(u1)
    add1 = Add()([l4,l2])

    u2 = UpSampling2D(size=2)(add1)
    l5 = Conv2D(64,kernel_size=3,activation="relu",padding="same")(u2)
    add2 = Add()([l5,l1])

    output = Conv2D(3,kernel_size=5,activation="linear",padding="same")(add2)
    
    model = Model(input_img,output)
    return model

model = Super_Resolution()
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
model.fit(train_lr,train_hr,epochs=100,batch_size=32,validation_data=(train_lr_val, train_hr_val),verbose=1)
```
